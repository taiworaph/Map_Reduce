```python
def NonNumericValuesOnly(LineRDD):
    """Function takes the RDD and subsets the RDD"""
    Values = LineRDD.split('\t')
    ZZ = []
    RealKey = str(Values[14]) + str(Values[15]) + str(Values[16])
    for ii in range(14,40):
        if Values[ii] == '':
            ZZ.append(str(0))
        else:
            ZZ.append(str(Values[ii]))
    
    return (RealKey, ZZ)
```
